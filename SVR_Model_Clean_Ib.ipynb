# Importing Dataset


```python
DF=pd.read_csv(path_in+'/All_Data_Final.csv')

#DF=DF.sample(frac=1)

y=DF['Ground Truth BPM']

X=DF.drop(columns='Ground Truth BPM')
```

```python
Col_R=[f'Data#_{i}_R' for i in range(150)] 

Col_C=[f'Data#_{i}_C' for i in range(150)] 

Col_F=[f'Data#_{i}_F' for i in range(150)] 
```

```python

y=DF['Ground Truth BPM']

X_R=DF.drop(columns='Ground Truth BPM')[Col_R]
X_C=DF.drop(columns='Ground Truth BPM')[Col_C]
X_F=DF.drop(columns='Ground Truth BPM')[Col_F]
```

```python
## Normalizing the Input Features and Target Variable
```

```python
from sklearn import preprocessing

X_Rs= pd.DataFrame(X_R)
min_max_scaler_R = preprocessing.MinMaxScaler()
x_R_scaled = min_max_scaler_R.fit_transform(X_Rs)
DF_R_All_norm = pd.DataFrame(x_R_scaled)
DF_R_All_norm.columns=[f'R_{i}' for i in range(150)]

X_Cs= pd.DataFrame(X_C)
min_max_scaler_C = preprocessing.MinMaxScaler()
x_C_scaled = min_max_scaler_C.fit_transform(X_Cs)
DF_C_All_norm = pd.DataFrame(x_C_scaled)
DF_C_All_norm.columns=[f'C_{i}' for i in range(150)]

X_Fs= pd.DataFrame(X_F)
min_max_scaler_F = preprocessing.MinMaxScaler()
x_F_scaled = min_max_scaler_F.fit_transform(X_Fs)
DF_F_All_norm = pd.DataFrame(x_F_scaled)
DF_F_All_norm.columns=[f'F_{i}' for i in range(150)]
```

```python
y_s= pd.DataFrame(y)

min_max_scaler_N= preprocessing.MinMaxScaler()
y_scaled_N= min_max_scaler_N.fit_transform(y_s)
y_all_norm = pd.DataFrame(y_scaled_N)
y_all_norm.columns=['Ground Truth BPM']
```

```python
DF_All_norm=pd.concat([y_all_norm,DF_R_All_norm,DF_C_All_norm,DF_F_All_norm], axis=1)
```

```python
DF_All_norm.to_csv('All_Data_Normalized.csv',index=False)
```

```python
DF_All_norm.drop(columns='Ground Truth BPM')
```

```python
from sklearn.model_selection import train_test_split

#X_norm=DF_All_norm.drop(columns='Ground Truth BPM')
X_norm=DF_R_All_norm.copy()
Y_norm=y_all_norm
X_train, X_test, y_train, y_test=train_test_split(X_norm,Y_norm,test_size=0.2)
```

```python
Y_norm
```

```python
X_norm
```

```python
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)

SVR_Model = svr_rbf.fit(X_train, y_train.values.reshape(-1,))
```

```python
y_pred_norm=svr_rbf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y_test_all=min_max_scaler_N.inverse_transform(y_test.values)

y_pred=np.array(svr_rbf.predict(X_test)).reshape(-1,1)

y_pred_all=min_max_scaler_N.inverse_transform(y_pred)
print('MAE=',mean_absolute_error(y_test_all,y_pred_all))
print('MSE=', mean_squared_error(y_test_all,y_pred_all))
```
